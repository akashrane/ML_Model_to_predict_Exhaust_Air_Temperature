```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras
```

```python
data = pd.read_excel("/content/trainingDataANN.xlsx")
```

```python
x_data = data[['hot air temperature', 'wet bulb temperature', 'expected moisture']]
y_data = data['exhaust air temperature']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2, random_state=42)
```

```python
model = keras.Sequential()
model.add(keras.layers.Dense(10, input_shape=(3,), activation='relu'))
model.add(keras.layers.Dense(1))
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.fit(x_train, y_train, epochs=500)
```

```python
test_loss = model.evaluate(x_test, y_test, verbose=2)
print('Test loss:', test_loss)
```

```python
column1 = float(input("Enter the value for hot air temperature : "))
column2 = float(input("Enter the value for wet bulb temperature: "))
column3 = float(input("Enter the value for expected moisture: "))

input_data = np.array([[column1, column2, column3]])

predictions = model.predict(input_data)

print("Predicted output: ", predictions[0][0])
```
